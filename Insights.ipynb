## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions: 
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Join Medicare and Diagnosis datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    medicare_df,
    parta_diagnosis_code_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Since Claims Header dataset has some principal diagnosis codes and the Diagnosis dataset 
#  supplements them with additional codes wherever possible, 
#  coalesce them with preference to the principal code from Claim Header dataset
# Once the diagnosis codes are combined into one column, remove the older columns and any duplicates
medicare_df['diagnosis_code'] = medicare_df[['prncpl_dgns_cd', 'clm_dgns_cd']].bfill(axis=1).iloc[:, 0]
medicare_df = medicare_df.drop(['prncpl_dgns_cd', 'clm_dgns_cd'], axis=1).drop_duplicates()
medicare_df.sort_values(by='claim_id').head(20)
```

```python
# Step 1: Count claims per HCP per year
hcp_claims_per_year = (
    cardiac_uniq_df
    .groupby(['npi_id', 'claim_year'])['claim_id']
    .nunique()
    .reset_index(name='claim_count')
)

# Step 2: Dynamically define bins that are always increasing
max_claims = hcp_claims_per_year['claim_count'].max()

# If the highest number of claims ≤ 100, push the last bin slightly higher
bins = [0, 1, 5, 10, 25, 50, 100, max_claims + 1 if max_claims <= 100 else max_claims]

# Ensure bins are strictly increasing (remove duplicates, sort)
bins = sorted(set(bins))

segment_labels = [
    '1 claim', '2–5 claims', '6–10 claims',
    '11–25 claims', '26–50 claims', '51–100 claims', '100+ claims'
]

# Assign segment labels safely
hcp_claims_per_year['hcp_segment'] = pd.cut(
    hcp_claims_per_year['claim_count'],
    bins=bins,
    labels=segment_labels[:len(bins)-1],  # trim to match bins
    include_lowest=True
)

display(hcp_claims_per_year.head())
print(f" Bins used: {bins}")
print(f" Max claim count in dataset: {max_claims}")
```

```python
# Step 3: Aggregate number of HCPs by year and segment
hcp_segments_summary = (
    hcp_claims_per_year
    .groupby(['claim_year', 'hcp_segment'])['npi_id']
    .nunique()
    .unstack(fill_value=0)
)

# Arrange segment order
hcp_segments_summary = hcp_segments_summary.reindex(columns=[c for c in segment_labels if c in hcp_segments_summary.columns])

print("🩺 HCP Segments per Year — Cardiac Claims (2016–2018)")
display(hcp_segments_summary)
```

```python
import pandas as pd

# 1️⃣ Ensure claim_year is numeric
cardiac_uniq_df['claim_year'] = pd.to_numeric(cardiac_uniq_df['claim_year'], errors='coerce')

# 2️⃣ Derive from claim_date if missing
if 'claim_year' not in cardiac_uniq_df.columns and 'claim_date' in cardiac_uniq_df.columns:
    cardiac_uniq_df['claim_year'] = pd.to_datetime(cardiac_uniq_df['claim_date'], errors='coerce').dt.year

# 3️⃣ Confirm valid years
cardiac_uniq_df = cardiac_uniq_df[cardiac_uniq_df['claim_year'].between(2016, 2018)]

# 4️⃣ Pick the correct patient-age column
age_col = 'age' if 'age' in cardiac_uniq_df.columns else 'patient_age'

# 5️⃣ Define age buckets
age_bins = [0, 59, 69, 79, 120]
age_labels = ['18–59', '60–69', '70–79', '80+']

cardiac_uniq_df['age_group'] = pd.cut(
    cardiac_uniq_df[age_col],
    bins=age_bins,
    labels=age_labels,
    include_lowest=True
)

print(" Age buckets assigned successfully.")
display(cardiac_uniq_df[['claim_year', age_col, 'age_group']].head())
```

```python
# Aggregate number of unique claims by claim_year and age_group
cardiac_age_claims = (
    cardiac_uniq_df
    .groupby(['claim_year', 'age_group'], observed=False)['claim_id']
    .nunique()
    .reset_index(name='claim_count')
    .sort_values(['claim_year', 'age_group'])
)

print("📊 Cardiac Claim Counts by Year & Age Group")
display(cardiac_age_claims)
```
